# Email traffic overtime

```
#Convert to dataframe
import pandas as pd
message_df = pd.read_sql_query(sql,conn)
message_df.info()
```

```
#sqlquery for number count of emails per day
sql = """

SELECT strftime('%Y/%d',date)
AS date,
COUNT (*) AS count FROM message
GROUP by date

"""
#Convert to dataframe
messageperday_df = pd.read_sql_query(sql,conn)
messageperday_df
```

```
#sqlquery for number count of emails per month
sql = """

SELECT strftime ('%Y-%m', date)
AS date,
COUNT(*) AS count FROM message
GROUP by date

"""
#Convert to dataframe
messagepermonth_df = pd.read_sql_query(sql,conn)
messagepermonth_df
```

```
#sql query for number count of emails per week
sql = """

SELECT strftime('%Y-%w',date)
AS date,
COUNT (*) AS count FROM message
GROUP BY date

"""
#Convert to dataframe
messageperweek_df = pd.read_sql_query(sql,conn)
messageperweek_df
```

#Top senders and recipients



```
#write sql query for top sender using count()
sql_sender = """
SELECT sender,COUNT(*)
AS count
FROM message
Group by sender
order by count
desc
Limit 15
"""



```

```
#Convert data to dataframe

Topsenders_df = pd.read_sql_query(sql_sender,conn)
Topsenders_df
```

```
#Display dataframe
print(Topsenders_df)
```

```
#Using the barplot
import matplotlib.pyplot
import seaborn as sns
data = Topsenders_df
sns.barplot(x="count",y="sender",data=data)
plt.show()
```

```
#write sql query for top recipient using count()
sql_recipient = """
SELECT rvalue,COUNT(*)
AS count
FROM Recipientinfo
Group by rvalue
order by count
desc
Limit 15
"""
```

```
#Convert data to dataframe

Toprecipients_df = pd.read_sql_query(sql_recipient,conn)
Toprecipients_df
```

```
#Using barplot
import matplotlib.pyplot
import seaborn as sns
data2 = Toprecipients_df
sns.barplot(x="count",y="rvalue",data=data2)
plt.show()
```

# Email Distribution By Recipient Type

```
#sql query to categorize by rtype using count
sql_rtype = """
SELECT rtype, COUNT(*) AS count
FROM Recipientinfo
GROUP BY rtype;
"""

```

```
#Convert data in dataframe
# Execute the query and fetch the results
result = conn.execute(sql_rtype).fetchall()

#Convert the query result into a dataframe
df = pd.DataFrame(result, columns=['rtype', 'count'])

#Close the database connection
conn.close()

#Print the dataframe
print(df)
```

```

#Plot graph with pie() and show()
plt.pie(df['count'], labels=df['rtype'], autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
